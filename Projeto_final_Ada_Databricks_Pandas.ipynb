```python
#Selecionando as colunas de interesse
tempo = datetime.now()
pd_columns_to_keep = [
    "Usage",
    "Account #",
    "Premise Code", 
    "Meter ID",
    "Read Type",   
    "Hour of the Day", 
    "Day of the Week",
    "Day",
    "Month",
    "Year"]

print(datetime.now() - tempo)
```

```python
df_pd['Year'].info()
```

```python
#Transfromar colunas em int

convert_dict = {'Year': int,
                'Month': int,
                'Day': int}
 
df_pd = df_pd.astype(convert_dict)
print(df_pd.dtypes)
print(datetime.now() - antes_importa)
```

# Criando dummy




```python
# Cria variáveis dummy 
antes = datetime.now()
df_dummies = pd.get_dummies(df_pd['Day of the Week'])

# Concatena as variáveis dummy com o DataFrame original
df_pd = pd.concat([df_pd, df_dummies], axis=1)

df_dummies = pd.get_dummies(df_pd['Meter ID'])
# Concatena as variáveis dummy com o DataFrame original
df_pd = pd.concat([df_pd, df_dummies], axis=1)


df_dummies = pd.get_dummies(df_pd['Read Type'])
# Concatena as variáveis dummy com o DataFrame original
df_pd = pd.concat([df_pd, df_dummies], axis=1)

df_pd = df_pd.drop('Read Type', axis= 1)
df_pd = df_pd.drop('Day of the Week', axis= 1)
df_pd = df_pd.drop('Meter ID', axis= 1)


# Exibe o DataFrame resultante
print(datetime.now() - antes)

print(df_pd)
```

```python
#Verificando os Dummies
display(df_pd.dtypes)
```

#Regressão Linear

```python
#Separando os dados (Retirando o target)
x = df_pd.drop(['Usage'], axis = 1)
y = df_pd["Usage"]
```

```python
#Separando os dados de Treino (80%) e Teste (20%)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

#Criando um Objeto de Regressão Linear
lr = LinearRegression()

#Treinando o Modelo
lr.fit(x_train,y_train)

#Verificando o R^2
r_sq = lr.score(x, y)
print('Coeficiente de Determinação (R²):', r_sq)

print(datetime.now() - antes_importa)
```
